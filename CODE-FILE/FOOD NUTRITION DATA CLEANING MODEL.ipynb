###### Drop rows with missing values

```python
df = df.dropna()
df
```

```python
df.shape
```

###### Fill missing values with the mean for numeric columns

```python
df = df.fillna(df.mean(numeric_only=True))
df
```

```python
dups = df.duplicated()
print('Number of duplicate rows = ',dups.sum())

df[dups]
```

##### Scatter Plot of Carbohydrate vs. Protein

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Data.Carbohydrate', y='Data.Protein', data=df)
plt.title('Carbohydrate vs. Protein')
plt.xlabel('Carbohydrate')
plt.ylabel('Protein')
plt.show()
```

```python
df.boxplot(column=['Data.Fat.Total Lipid'])
plt.show()
```

```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3=col.quantile([0.25,0.75])
    IQR=Q3-Q1
    lower_range= Q1-(1.5 * IQR)
    upper_range= Q3+(1.5 * IQR)
    return lower_range, upper_range    
```

```python
lrfat,urfat=remove_outlier(df['Data.Fat.Total Lipid'])
df['Data.Fat.Total Lipid']=np.where(df['Data.Fat.Total Lipid']>urfat,urfat,df['Data.Fat.Total Lipid'])
df['Data.Fat.Total Lipid']=np.where(df['Data.Fat.Total Lipid']<lrfat,lrfat,df['Data.Fat.Total Lipid'])
```

```python
df.boxplot(column=['Data.Fat.Total Lipid'])
plt.show()
```
